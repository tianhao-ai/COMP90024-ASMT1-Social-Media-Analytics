## Clustering and Clouding Computing project 1 

Aim of this project:
• count the number of different tweets made in the Greater Capital cities of Australia, 
• identify the Twitter accounts (users) that have made the most tweets, and 
• identify the users that have tweeted from the most different Greater Capital cities. 

```python
import pandas as pd
import json
```

```python
# Load data
sal_data = pd.read_json('sal.json').transpose()
twitter_data = pd.read_json('twitter-data-small.json')
```

```python
sal_data.head()
```

```python
sal_data = sal_data.reset_index()
```

```python
twitter_data['location']=twitter_data.apply(lambda x:x['includes']['places'][0]['full_name'].split(',')[0],axis=1)
```

```python
twitter_data['location'] = twitter_data['location'].apply(str.lower)
```

```python
twitter_data['full_location'] = twitter_data.apply(lambda x:x['includes']['places'][0]['full_name'].split(','),axis=1)
```

```python
new_data = pd.merge(twitter_data,sal_data,on='location')
```

```python
city_list = {}
with open("sal.json") as f:
    data = json.load(f)
    for key in data:
        location = key
        if data[key]['gcc'] in city_list:
            city_list[data[key]['gcc']].append(location)
        else:
            city_list[data[key]['gcc']]=[]
            city_list[data[key]['gcc']].append(location)
```
